### Calulating and Plotting the Housing Units per Year

Step 1: Using `groupby` function to group the data by year. Aggregating the results by the `mean` of the groups.

```python
# Creating a numerical aggregation that groups the data by the year and then averages the results.

housing_units_by_year = sfo_data_df.drop(columns=["sale_price_sqr_foot", "gross_rent"]).groupby("year").mean()

# Reviewing the DataFrame

housing_units_by_year
```

Step 2: Using the `hvplot` function to plot the `housing_units_by_year` DataFrame as a bar chart. The x-axis represents the `year` and the y-axis represents the `housing_units`.

Step 3: Styling and formatting the line plot to ensure a professionally styled visualization.

```python
# Creating a visual aggregation to explore the housing units by year

display(housing_units_by_year.hvplot.bar(
    x="year",
    y="housing_units",
    xlabel="Year",
    ylabel="Housing Units",
    rot=0,
    color = "Blue",
    yformatter = "%.0f",
    height = 450,
    width = 750,
    fontscale = 1.3,
    ylim = (364800, 385900),
    hover_color = "Grey",
    title="Housing Units in San Francisco from 2010 to 2016"
))
```

### Comparing the Average Sale Prices by Neighborhood

Step 1: Creating a new DataFrame that groups the original DataFrame by year and neighborhood. Aggregating the results by the `mean` of the groups.

```python
# Grouping by year and neighborhood and then creating a new dataframe of the mean values

prices_by_year_by_neighborhood = sfo_data_df.groupby(["year", "neighborhood"]).mean()

# Rewing the DataFrame

prices_by_year_by_neighborhood
```

Step 2: Filtering out the “housing_units” column to create a DataFrame that includes only the `sale_price_sqr_foot` and `gross_rent` averages per year.

```python
# Filtering out the housing_units

prices_by_year_by_neighborhood.drop(columns=["housing_units"], inplace=True)

# Reviewing the first and last five rows of the DataFrame

display(prices_by_year_by_neighborhood.head())

display(prices_by_year_by_neighborhood.tail())
```

Step 3: Creating an interactive line plot with hvPlot that visualizes both `sale_price_sqr_foot` and `gross_rent`. Setting the x-axis parameter to the year (`x="year"`). Using the `groupby` parameter to create an interactive widget for `neighborhood`.

Step 4: Styling and formatting the line plot to ensure a professionally styled visualization.

```python
# Using hvplot to create an interactive line plot of the average price per square foot
# The plot will have a dropdown selector for the neighborhood

display(prices_by_year_by_neighborhood.hvplot.line(
    x = "year",
    xlabel = "Year",
    ylabel = "Gross Rent / Sale Price Per Square Foot",
    hover = True,
    hover_color = "Aqua",
    fontscale = 1.3,
    height = 400,
    frame_width = 590,
    groupby = "neighborhood",
    title="Sale Price Per Square Foot and Average Gross Rent - 2010-2016 - By Neighborhood"
))
```

### Building an Interactive Neighborhood Map

Step 1: Reading the `neighborhood_coordinates.csv` file from the `Resources` folder into the notebook, and creating a DataFrame named `neighborhood_locations_df`. Making sure to set the `index_col` of the DataFrame as “Neighborhood”.

```python
# Loading neighborhoods coordinates data

neighborhoods_coordinates_path = Path(r"C:\Users\prath\UofT\UTOR-VIRT-FIN-PT-08-2023-U-LOLC\06-PyViz\Unit 6 Homework\Instructions\Starter_Code\Resources\neighborhoods_coordinates.csv")

neighborhood_locations_df = pd.read_csv(neighborhoods_coordinates_path, index_col="Neighborhood")

# Reviewing the DataFrame

neighborhood_locations_df
```

Step 2: Using the original `sfo_data_df` Dataframe, creating a DataFrame named `all_neighborhood_info_df` that groups the data by neighborhood. Aggregating the results by the `mean` of the group.

```python
# Calculating the mean values for each neighborhood

all_neighborhood_info_df = sfo_data_df.drop(columns=["year"]).groupby(["neighborhood"]).mean()

# Reviewing the resulting DataFrame

all_neighborhood_info_df
```

Step 3: Concatenating the `neighborhood_locations_df` DataFrame with the `all_neighborhood_info_df` DataFrame into `all_neighborhoods_df`.

```python
# Using the Pandas concat function to join the 
# neighborhood_locations_df and the all_neighborhood_info_df DataFrame

all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighborhood_info_df], 
    axis=1,
    join="inner",
    sort=False
)

# Rounding the values of sale_price_sqr_foot and gross_rent

all_neighborhoods_df = round(all_neighborhoods_df, 4)

# Reviewing the DataFrame

display(all_neighborhoods_df.head())

display(all_neighborhoods_df.tail())
```

```python
# Calling the dropna function to remove any neighborhoods that do not have data

all_neighborhoods_df = all_neighborhoods_df.dropna().reset_index()

# Renaming the "index" column as "Neighborhood" for use in the Visualization

all_neighborhoods_df = all_neighborhoods_df.rename(columns={
    "index": "Neighborhood"})

# Reviewing the resulting DataFrame

display(all_neighborhoods_df.head())

display(all_neighborhoods_df.tail())
```

Step 4: Using hvPlot with GeoViews enabled, creating a `points` plot for the `all_neighborhoods_df` DataFrame. Also making sure to:

* Set the `geo` parameter to True.

* Set the `size` parameter to "sale_price_sqr_foot".

* Set the `color` parameter to "gross_rent".

* Set the `tiles` parameter to "OSM".

* Set the `alpha` parameter to "0.75".

* Set the `hover_cols` parameter to "all".

* Set the `use_index` parameter to "False".

* Set the `frame_width` parameter to 700.

* Set the `frame_height` parameter to 500.

* Finally, include a descriptive title.

```python
# Creating a plot to analyze neighborhood info

display(all_neighborhoods_df.hvplot.points(
    "Lon",
    "Lat",
    geo = True,
    size = "sale_price_sqr_foot",
    color = "gross_rent",
    tiles = 'OSM',
    hover_cols = 'all',
    use_index = False,
    alpha = 0.75,
    frame_width = 700,
    frame_height = 500,
    title = "Geographic View of Aggregate of Sale Price Per Square Foot and Gross Rent - 2010-2016 - San Francisco"
))
```
